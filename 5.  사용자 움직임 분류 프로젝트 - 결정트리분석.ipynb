```python
### features.txt 파일에 있는 컬럼명을 입력 받아서 중복된 컬럼명은 원본 컬럼명+_1, _2와 같이 중복된 차수를 원본 컬럼명에 더해서 컬럼명을 update 하는 함수임. . 
def get_new_feature_name_df(old_feature_name_df):
    #column_name으로 중복된 컬럼명에 대해서는 중복 차수 부여, col1, col1과 같이 2개의 중복 컬럼이 있을 경우 1, 2 
    feature_dup_df = pd.DataFrame(data=old_feature_name_df.groupby('feature_name').cumcount(), columns=['dup_cnt'])
    # feature_dup_df의 index인 column_name을 reset_index()를 이용하여 컬럼으로 변환. 
    feature_dup_df = feature_dup_df.reset_index()
    # 인자로 받은 features_txt의 컬럼명 DataFrame과 feature_dup_df를 조인. 
    new_feature_name_df = pd.merge(old_feature_name_df.reset_index(), feature_dup_df, how='outer')
    # 새로운 컬럼명은 앞에 중복 차수를 접미어로 결합. 
    new_feature_name_df['feature_name'] = new_feature_name_df[['feature_name', 'dup_cnt']].apply(lambda x : x[0]+'_'+str(x[1]) 
                                                                                           if x[1] >0 else x[0] ,  axis=1)
    new_feature_name_df = new_feature_name_df.drop(['index'], axis=1)
    return new_feature_name_df
```

```python
# 결정트리 모델 
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_HAR = DecisionTreeClassifier(random_state = 156)  # 결정트리분류 모델 생성
```

```python
dt_HAR.fit(X_train,Y_train)
X_test = X_test.dropna(axis=1)
Y_predict = dt_HAR.predict(X_test)
Y_predict
```

```python
# 결과분석
from sklearn.metrics import accuracy_score
```

```python
# 정확도 분석
accuracy = accuracy_score(Y_test,Y_predict)
print('결정 트리 예측 정확도 : {0:.4f}'.format(accuracy))
```

```python
# 최적의 파라미터 찾기
#파라미터로 정확도를 올릴수있다?
print("결정 트리의 현재 하이퍼 파라미터 : \n",dt_HAR.get_params())
```

```python
from sklearn.model_selection import GridSearchCV
# 최적의 파라미터를 찾게 해준다?
```

```python
params = {
    #트리의 깊이를 설정
    'max_depth' : [ 6, 8, 10, 12, 16, 20, 24]
}

grid_cv = GridSearchCV(dt_HAR, param_grid=params, scoring='accuracy', 
                       cv=5, return_train_score=True)
grid_cv.fit(X_train , Y_train)
```

```python
cv_results_df = pd.DataFrame(grid_cv.cv_results_)
cv_results_df[['param_max_depth', 'mean_test_score','mean_train_score']]
```

```python
print('최고 평균 정확도 : {0:.4f}, 최적 하이퍼 파라미터 :{1}'.format(grid_cv.best_score_, grid_cv.best_params_))
```
