```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import numpy as np
import pandas as pd
data_df = pd.read_csv('./10장_data/auto-mpg.csv',header = 0, engine = 'python')
```

```python
print('데이터셋 크기:', data_df.shape)
```

```python
data_df.head()
```

```python
#분석하지 않을 변수 제외하기
data_df = data_df.drop(['car_name','origin','horsepower'], axis = 1, inplace = False)
```

```python
print('데이터세트 크기 : ', data_df.shape)
data_df.head()
```

```python
data_df.info()
```

```python
#분석 모델 구축
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
Y = data_df['mpg']
X = data_df.drop(['mpg'],axis = 1, inplace = False)
```

```python
X_train, X_test,Y_train, Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)  # 선형회귀분석 모델 훈련시키기
Y_predict = lr.predict(X_test) # 평가 데이터에 대한 예측 수행
print(Y_predict)
```

# 결과 분석 및 시각화


```python
mse = mean_squared_error(Y_test,Y_predict)
rmse = np.sqrt(mse)
```

```python
# 오차로 평가지표 구하기
print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y절편 값 : ', np.round(lr.intercept_,2))
print('회귀 계수 값:', np.round(lr.coef_,2))  # 기울기
```

```python
coef = pd.Series(data=np.round(lr.coef_,2),index=X.columns)
coef.sort_values(ascending = False)
```
