```python
all_files_data = [] #저장할 리스트 

for file in all_files:
    data_frame = pd.read_excel(file)
    all_files_data.append(data_frame)

all_files_data[0] #출력하여 내용 확인
```

### 1-2. 데이터 전처리 (Pre-processing)

```python
# 제목 추출
all_title = all_files_data_concat['제목']

all_title #출력하여 내용 확인
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
# NLTK 를 통해서 불용어 제거하기
example = "Family is not an important thing. It's everything."
stop_words = set(stopwords.words('english')) 

word_tokens = word_tokenize(example)

result = []
for w in word_tokens: 
    if w not in stop_words: #불용어가 아닌 단어들만 리스트에 추가
        result.append(w) 

print(word_tokens) 
print('불용어 제거후:', result) 
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
words = ['policy', 'doing', 'organization', 'have', 'going', 'love', 'lives', 'fly', 'dies', 'watched', 'has', 'starting']
print([lemmatizer.lemmatize(w) for w in words])
```

```python
stopWords = set(stopwords.words("english"))
lemma = WordNetLemmatizer()
```

```python
words = []  

for title in all_title:
    #알파벳이 아닌 단어는 공백으로 치환, 소문자 정규화, 단어 토근화,불용어 제거
    EnWords = re.sub(r"[^a-zA-Z]+", " ", str(title))    
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]
    words.append(EnWordsTokenStopLemma)
```

```python
print(words)  #출력하여 내용 확인
```

```python
words2 = list(reduce(lambda x, y: x+y,words)) # 2차원 리스트를 1차원으로 바꾸기
print(words2)  #작업 내용 확인
```

# 2. 데이터 탐색

## 2-1. 단어 빈도 탐색

```python
count = Counter(words2)

count   #출력하여 내용 확인
```

```python
word_count = dict()

#출력횟수많은 상위 50개 단어

for tag, counts in count.most_common(50):  
    if(len(str(tag))>1):  #단어 길이 1보다 큰거
        word_count[tag] = counts
        print("%s : %d" % (tag, counts))
```

#### 여기서 잠깐!! : 검색어로 사용한 big'과 'data' 빈도가 압도적으로 많으므로, 이를 제거한다.

```python
#검색어로 사용한 'big'과 'data' 항목 제거 하기
del word_count['big']
del word_count['data']
```

## 2-2 단어 빈도 히스토그램

```python
# 히스토그램 표시 옵션 
plt.figure(figsize=(12,5))
plt.xlabel("word")
plt.ylabel("count")
plt.grid(True)

sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)
print(sorted_Keys)
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation='85')

plt.show()
```
