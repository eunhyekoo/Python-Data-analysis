```python
from scipy import stats
from statsmodels.formula.api import ols,glm
```

```python
# wine.loc[행 범위, '열 컬럼']
red_wine_quality = wine.loc[wine['type'] == 'red','quality']
white_wine_quality = wine.loc[wine['type'] == 'red', 'quality']
print(white_wine_quality)
```

```python
red_wine_quality = wine.loc[wine['type'] == 'red','quality']
white_wine_quality = wine.loc[wine['type'] == 'white','quality']
print(red_wine_quality)
```

```python
# t-검정을 하고 두 그룹간의 차이를 확인
result = stats.ttest_ind(red_wine_quality, white_wine_quality, equal_var = False)
print('t statistic : %.3f \np-value : %.3f' % (result))
```

```python
Rformula = '''quality ~ fixed_acidity + volatile_acidity + citric_acid + 
            residual_sugar + chlorides + free_sulfur_dioxide + total_sulfur_dioxide +
            density + pH + sulphates + alcohol'''

# 종속변수와 독립변수 구성
```

```python
# 선형회귀 모델 중에서 OLSOrdinary Least Squars 모델 사용
regression_result = ols(Rformula, data = wine).fit()
regression_result
```

```python
# 선형회귀 모델 중에서 OLSOrdinary Least Squars 모델 사용
regression_result = ols(Rformula, data = wine).fit()
# 통계값 출력
regression_result.summary()
```

## 부분회귀 플롯으로 시각화하기

```python
import statsmodels.api as sm

others = list(set(wine.columns).difference(set(["quality","fixed_acidity"])))
#fixed_acidity 값이 quality에 미치는 영향에 부분 회귀를 수행
p,resids = sm.graphics.plot_partregress("quality", "fixed_acidity", others, data = wine, ret_coords = True)
#종속변수, 독립변수, 독립변수를 제외한 나머지 독립변수 리스트,
plt.show()

fig = plt.figure(figsize = (8,13))
# 다중 선형 회구 분석 결과를 가지고 있는 regression_result를 이용해서 부분 회귀 플롯 구함
sm.graphics.plot_partregress_grid(regression_result, fig = fig)
plt.show()
```
